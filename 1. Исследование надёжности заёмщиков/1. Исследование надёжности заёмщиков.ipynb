### Лемматизация

```python
#Создадим новую переменную, в которой сохраним данные столбца с целями кредита
credit_purposes = statistics['purpose'].unique().tolist()
#Проверим результат
credit_purposes
```

```python
#Переведем данные в строку
from pymystem3 import Mystem
m = Mystem()
string = '; '.join(credit_purposes)
(string)
```

```python
#Проведем лемматизацию значений
lemmas = m.lemmatize(string)
from collections import Counter
(Counter(lemmas))
```

*Используя выделенные леммы существительных можно определить следующие основные цели кредита:*
1) Недвижимость/ жилье
2) Автомобиль
3) Образование
4) Свадьба

```python
# Создадим список новых категорий по цели кредита
categories = ["недвижимость", "жилье", "автомобиль", "образование", "свадьба"]

# Проведем лемматизацию,
# одновременно заменим список лемм в каждой строке на главное ключевое слово из списка категорий

def lemmatize(text):
    lemma = m.lemmatize(text)
    for word in categories:
        if word in lemma:
            lemma = word
    return lemma
# Создадим новый столбец с категориями по целям кредита
statistics['purpose_group'] = statistics['purpose'].apply(lemmatize)        
statistics['purpose_group'].value_counts().head(15)
```

```python
#Так как жилье относится к недвижимости, сделаем соответствующую замену
statistics.loc[statistics['purpose_group'] == 'жилье', 'purpose_group'] = 'недвижимость'
#Проверим
statistics['purpose_group'].value_counts()
```

**Вывод**
    
    Используя выделенные леммы существительных определили следующие основные цели кредита:
    1) недвижимость (включая жилье)
    2) автомобиль
    3) образование
    4) свадьба

### Категоризация данных

    Мы уже категоризировали цели кредита. Исходя из цели исследования, проведем категоризацию следующих данных:
- количество детей (для решения задачи исследования можно выделить всего две категории: бездетные, с детьми (ведь нас интересует влияет ли *наличие* детей на своевременность возврата кредита);
- уровень дохода (воспользуемся стандартным подходом, используем процентели в имеющихся данных: низкий 0-30%, средний 30-60%, выше среднего 60-90 %, высокий > 90 %; выборка порогов субъективна и однозначно условна);
- наличие задолженности по возврату кредитов (удобно видеть наглядно, у кого имеются просрочки платежей, очевидны две категории: должники, своевременный возврат кредита).
    
    Семейное положение категоризировать не будем, основные категории уже приведены в данных.

```python
#Используем функцию для создания категорий по количеству детей
def children_status(children):
    if children > 0: return 'С детьми'
    else: return 'Бездетные'

#Применим метод, добавим новый столбец и проверим результат
statistics['children_status'] = statistics['children'].apply(children_status)
statistics['children_status'].value_counts()
```

Данные по детям категоризированы.

```python
#Определим процентили столбца с уровнем дохода, воспользуемся quantile()
(statistics['total_income'].quantile([0.30,0.60,0.90]))
```

```python
#Используем функцию для создания категорий по уровню дохода
def income_status(total_income):
    if total_income <= 116062.0:
            return 'Низкий уровень дохода'
    if total_income <= 156400.0:
            return 'Средний уровень дохода'
    if total_income <= 269784.0:
            return 'Выше среднего уровень дохода'
    return 'Высокий уровень дохода'
#Применим метод, добавим новый столбец и проверим результат
statistics['income_status'] = statistics['total_income'].apply(income_status)
statistics['income_status'].value_counts()
```

Данные по доходу категоризированы.

```python
#Используем функцию для создания категорий по своевременности возврата кредита
def debt_status(debt):
    if debt > 0: return 'Должники'
    else: return 'Своевременный возврат кредита'

#Применим метод, добавим новый столбец и проверим результат
statistics['debt_status'] = statistics['debt'].apply(debt_status)
statistics['debt_status'].value_counts()
```

Данные по своевременности возврата кредита категоризированы

## Шаг 3. Ответьте на вопросы

- **Есть ли зависимость между наличием детей и возвратом кредита в срок?**

```python
#Для наглядности удобно создать сводную таблицу, используем метод pivot_table()
children_pivot = statistics.pivot_table(index = ['children_status'], columns = ['debt'], values = 'debt_status', aggfunc='count')
#Создадим новый столбец, в котором отобразим процент должников по каждой категории,
#для большинства людей результат в процентах легче оценить, чем сравнивать отношение
children_pivot['debt_%'] = children_pivot[1] / (children_pivot[0] + children_pivot[1]) * 100
#Так как нас интересую в первую очередь неблагонадежные клиенты, отсортируем по убыванию
children_pivot.sort_values(by = 'debt_%', ascending = False)
```

**Вывод**: В целом задолженность по выплате кредитов среди клиентов с детьми и бездетных находится примерно на одном уровне. Однако почти на 2% больше клиентов с детьми имеют задолженность по выплате кредита, что можно объяснить расстановкой приоритетов в социуме - в случае непредвиденных обстоятельств обеспечить насущные потребности детей важнее, чем вовремя погасить кредит.А также статистически не верной выборкой - в исходных данных в два раза больше бездетных клиентов.
В тоже время, наличие или отсутствие детей не может быть основным критерием благонадежности клиента. Количество детей по сути необходимо знать только для расчета приблизительного уровня расходов клиента.

- **Есть ли зависимость между семейным положением и возвратом кредита в срок?**

```python
#Для наглядности удобно создать сводную таблицу, используем метод pivot_table()
family_status_pivot = statistics.pivot_table(index = ['family_status'], columns = ['debt'], values = 'debt_status', aggfunc='count')
#Создадим новый столбец, в котором отобразим процент должников по каждой категории,
#для большинства людей результат в процентах легче оценить, чем сравнивать отношение
family_status_pivot['debt_%'] = family_status_pivot[1] / (family_status_pivot[0] + family_status_pivot[1]) * 100
#Так как нас интересую в первую очередь неблагонадежные клиенты, отсортируем по убыванию
family_status_pivot.sort_values(by = 'debt_%', ascending = False)
```

**Вывод**: Неплательщики имеются среди людей всех семейных статусов. Однако среди юридически свободных от уз брака людей (не женат/не замужем, гражданский брак) чаще встречаются просрочки платежей (разница примерно в 2 % клиентов). Опыт семейной жизни (женат/замужем, в разводе, вдовец/вдова) вырабатывает в людях ответственность и чувство долга, а также умение правильно планировать свой бюджет, поэтому такие клиенты реже попадают в категорию должников. 
Несмотря на незначительную разницу между количеством должников в разных категориях, все же учитывать семейный статус человека при  оценке благонадежности клиента следует, ведь социальные риски сильнее проявляются на одиноких/свободных людях. Отсутствие обязательств в такой социально значимой сфере как брак откладывает отпечаток на отношение к обязательствам в других сферах жизни.

- **Есть ли зависимость между уровнем дохода и возвратом кредита в срок?**

```python
#Для наглядности удобно создать сводную таблицу, используем метод pivot_table()
income_status_pivot = statistics.pivot_table(index = ['income_status'], columns = ['debt'], values = 'debt_status', aggfunc='count')
#Создадим новый столбец, в котором отобразим процент должников по каждой категории,
#для большинства людей результат в процентах легче оценить, чем сравнивать отношение
income_status_pivot['debt_%'] = income_status_pivot[1] / (income_status_pivot[0] + income_status_pivot[1]) * 100
#Так как нас интересую в первую очередь неблагонадежные клиенты, отсортируем по убыванию
income_status_pivot.sort_values(by = 'debt_%', ascending = False)
```

Это можно объяснить тем, что данная категория клиентов наиболее социально стабильна, с одной стороны они обеспечены и защищены от наиболее распростаненных рисков, с другой - не играют "по - крупному", не входят в категорию высокообеспеченных людей, в которой действуют свои законы и свои риски. 
Клиенты со средним уровнем дохода чаще являются должниками, однако разница с должниками среди клиентов с относительно низким уровнем дохода незначительна, вероятно это связано с относительным характером выборки  для данных категорий.
Уровень доходов является основным критерием для оценки платежеспособности клиента. Однако благонадежность клиента зависит в первую очередь не от дохода, а от базовых качеств его личности и жизненных обстоятельств.
Нельзя аргументировано утверждать, что уровень дохода влияет на его надежность как клиента банка, уровень дохода влияет только на вероятную возможность оплатить кредит вовремя.

Данная цель однозначно в зоне риска по благонадежности клиента.

Кредит на свадьбу своевременно оплачивают чаще, чем кредиты на автомобиль и образование, вероятно это связано с относительно небольшими суммами кредита.
