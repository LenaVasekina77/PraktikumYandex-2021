## 2. Анализ данных

### 2.1. Определение ТОП-10 городов по количеству рейсов

```python
city_top = city.sort_values('average_flights', ascending=False).head(10).reset_index(drop = True)
city_top
```

### 2.2. Графики 

#### 2.2.1. Распределение количества рейсов по моделям самолетов

```python
# Построим гистограмму распределения рейсов по моделям самолетов
models.sort_values(by='flights_amount', ascending=True).plot(x='model', y='flights_amount',
    kind='barh',figsize=(15, 6), fontsize=15, color='b', ec='black')
plt.title('Распределение рейсов по моделям самолетов в сентябре 2018 года', fontsize=15)
plt.legend(prop={'size': 16}, title = 'Количество рейсов')
plt.xlabel('Количество рейсов, шт')
plt.ylabel('Модель самолета')
```

**ВЫВОД:**

**Лидирами по числу рейсов в сентябре 2018 года являются три модели самолетов: Cessna 208 Caravan, Bombardier CRJ-200, Sukhoi SuperJet-100, вероятно, количество данных моделей в парке самолетов авиакомпании наибольшее. Учитывая, что авиакомпания "F9" занимается внутренними пассажирскими перевозками, неудивительно, что в парке самолетов отдано предпочтение небольшим по пассажировместимости моделям.**

**Cessna 208 Caravan** - используется на авиалиниях малой протяженности, перевозит до 9 пассажиров - https://ru.wikipedia.org/wiki/Cessna_208

**Bombardier CRJ-200** - региональный самолет, способен выполнять полёты в сложных метеорологических условиях и в условиях высокогорных аэродромов, 50-местная модель - https://ru.wikipedia.org/wiki/Bombardier_CRJ200

**Sukhoi SuperJet-100** -  ближнемагистральный самолёт, предназначенный для перевозки от 87 до 108 пассажиров на дальность 3050 или 4600 км - https://ru.wikipedia.org/wiki/Sukhoi_Superjet_100


#### 2.2.2. Распределение количества рейсов по городам прилета

```python
# Построим гистограмму распределения рейсов по городам прилета
city.sort_values(by='average_flights', ascending=True).plot(x='city', y='average_flights',
    kind='barh',figsize=(15, 50), fontsize=15, color='g', ec='black')
plt.title('Распределение среднедневного количества рейсов по городам прилета в августе 2018 года', fontsize=15)
plt.legend(prop={'size': 16}, title = 'Количество рейсов в день, в среднем')
plt.xlabel('Среднедневное количество рейсов, шт')
plt.ylabel('Город прилета') 
```

**ВЫВОД:**

**Авиакомпания выполняет внутренние перевозки практически во всех регионах России. Наибольшее количество рейсов совершается в Москву, заметно меньше - в Санкт- Петербург, еще меньше в Новосибирск, Красноярск и Екатеринбург входят в пятерку лидеров. Около 20 городов ежедневно принимают от 5 до 10 рейсов, в остальные города ежедневно летает менее 5 самолетов авиакомпании "F9".** 

#### 2.2.3. Распределение рейсов в ТОП-10 популярных городов

```python
# Построим гистограмму распределения рейсов по ТОП-10 городов прилета
city_top.sort_values(by='average_flights', ascending=True).plot(x='city', y='average_flights',
    kind='barh',figsize=(15, 6), fontsize=15, color='y', ec='black')
plt.title('Распределение среднедневного количества рейсов в ТОП-10 городов прилета в августе 2018 года', fontsize=15)
plt.legend(prop={'size': 16}, title = 'Количество рейсов в день, в среднем')
plt.xlabel('Среднедневное количество рейсов, шт')
plt.ylabel('Город прилета') 
```

**ВЫВОД:**

**В ТОП-10 лидеров по рейсам  авиакомпании "F9" входит Москва (около 130 рейсов в день в августе 2018 года), Санкт- Петербург (чуть более 30 рейсов в день в августе 2018 года), Новосибирск (17 рейсов), Красноярск и Екатеринбург (около 11 рейсов) входят в пятерку лидеров. Остальные лидеры ТОП-10 ежедневно около 10 рейсов авиакомпании "F9".**
