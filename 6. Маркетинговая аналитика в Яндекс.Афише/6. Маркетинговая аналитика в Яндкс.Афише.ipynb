## 3. Продуктовые метрики

```python
# Для удобства работы с данными, выделим в отдельные столбцы дату начала сессии, недели, месяца и года
visits['session_date'] = visits['start_ts'].dt.date
visits['session_week'] = visits['start_ts'].astype('datetime64[W]')
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')
visits['session_year'] = visits['start_ts'].astype('datetime64[Y]')
visits.head()
```

### 3.1. DAU, WAU, MAU

```python
# Посчитаем метрики. Сгруппируем данные по уникальным пользователям, дате, неделе, месяцу.
dau_total = visits.groupby('session_date').agg({'uid': 'nunique'})
wau_total = visits.groupby('session_week').agg({'uid': 'nunique'})
mau_total = visits.groupby('session_month').agg({'uid': 'nunique'})

print('Средний DAU за весь период:', int(dau_total.mean()), 'пользователей в день')
print('Средний WAU за весь период:', int(wau_total.mean()), 'пользователей в неделю')
print('Средний MAU за весь период:', int(mau_total.mean()), 'пользователей в месяц')
```

```python
# Рассчитаем Sticky Factor = DAU / MAU * 100%
sticky_factor =  dau_total / mau_total * 100
print('Средний Sticky Factor за весь период:', int(sticky_factor.mean()), '%')
```

**Sticky Factor имеет минимальное значение, которое может принимать данная метрика, и говорит оно о том, что пользователи не задерживаются на Яндекс.Афише.**

### 3.4. Retention Rate

```python
# Определим дату первого посещения, сгруппируем по пользователям 
first_visit = visits.groupby(['uid'])['start_ts'].min().reset_index()
first_visit.columns = ['uid', 'first_visit']
first_visit.head()
```

```python
# Объединим с исходным датафреймом visits.
visits = visits.merge(first_visit)
visits.head()
```

```python
# Изменим тип столбцов
visits['first_visit_month'] = visits['first_visit'].astype('datetime64[M]')
visits['visit_month'] = visits['start_ts'].astype('datetime64[M]')
```

```python
#Для каждой строки датафрейма рассчитаем время активности пользователя в рамках когорты, период - месяц
visits['visit_month_activity'] = (visits['visit_month'] - visits['first_visit_month']) / np.timedelta64(1, 'M')
visits['visit_month_activity'] = visits['visit_month_activity'].round().astype('int')
```

```python
# Создадим сводную таблицу по когорте и времени активности пользователя в рамках когорты 
# Посчитаем для каждой когорты количество активных пользователей на определённый месяц
cohorts = visits.pivot_table(index = 'first_visit_month',
                         columns = 'visit_month_activity',
                         values = 'uid',
                         aggfunc = 'nunique')

cohorts.fillna('')
```

```python
# Рассчитаем Retention rate
retention = pd.DataFrame()
for col in cohorts.columns:
    retention = pd.concat([retention, cohorts[col]/cohorts[0]], axis = 1)
retention.columns = cohorts.columns

retention.fillna('')
```

```python
#Построим тепловую карту
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Процент возвращающихся клиентов по когортам', size=14)
ax = sns.heatmap(retention, annot=True, fmt='.1%', cmap= 'BuGn', linewidths=3, linecolor='black', vmin=0, vmax=0.08, cbar=False)
ax.set_yticklabels(cohorts.index.date)
ax.set_ylabel('Месяц начала когорты')
ax.set_xlabel('Месяц жизни когорты')
```

```python
print('В среднем {0:.0%} пользователей возвращается на Афиша.Яндекс на следующий месяц после первого посещения.'.format(retention[1].mean()))
```

В среднем 7% пользователей возвращается на следующий месяц после первого использования сайта. На тепловой карте видно, что на второй месяц возвращается 2,4-6,8% пользователей. С каждым месяцем число удержанных пользователей медленно падает, только в первой когорте на 4, 5 и 11 месяцы количество возврашающихся пользователей повышается. В более "молодых" когортах retention rate заметно ниже, чем в 1,3,4,5 и 6 когортах, где наибольший показатель в первый месяц жизни когорты.

**Таким образом, наибольший процент удержания клиентов наблюдается в когортах июня, августа, сентября, октября и ноября 2017 года, из них наиболее стабильна когорта июня. Учитывая направления сервиса Яндекс.Афиши, можно предположить, что когорты постоянных посетителей сайта - фанатов театра, кино и других мероприятий, сформировались еще до июня 2017 года, а приток новых клиентов и их удержание связаны не только с маркетингом, но и с графиком проведения, спецификой и качеством афишируемых мероприятий, а также с огромным количеством внешних факторов (например, возраст клиента, насколько он легок на подъем, насколько финансово обеспечен, насколько занят и т.д. и т.п.).**

```python
# код ревьюера: переопределим индекс таблицы для тепловой карты:
retention.index = pd.to_datetime(retention.index).strftime('%Y-%m')

#Построим тепловую карту
plt.figure(figsize=(15, 5))
sns.heatmap(retention, 
            annot=True, 
            fmt='.1%', 
            linewidths=0.5, 
            linecolor='white',  
            vmax=0.1, cmap='Blues',
            ).set(title = 'Retention Rate')
plt.ylabel("Cohort") 
plt.xlabel("Cohort age (month)") 
plt.show()
```

### 4.4. LTV

```python
# В таблице orders_1 (уже содержит месяц первого заказа)
# Посчитаем количество новых покупателей (n_buyers) за каждый месяц
cohort_2 = (orders_1.groupby('first_order').agg({'uid': 'nunique'}).reset_index())
cohort_2.columns = ['first_order', 'n_buyers']
cohort_2.head() 
```

```python
# Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку 
cohorts_3 = (orders_1.groupby(['first_order', 'buy_ts']).agg({'revenue': 'sum'}).reset_index())
cohorts_3.head()
```

```python
# Рассчитаем усредненный ROMI для каждого источника, для этого:
# Заново сформируем таблицу с сумаррными данными расходов, числа клиентов и выручкидля каждого источника
# Рассчитаем усредненный LTV и CAC
costs_2 = costs.pivot_table(
        index= ['source_id'], 
        values='costs', 
        aggfunc='sum').reset_index()

n_buyers = n_buyers_per_source

revenue = orders.groupby(['source_id']).agg({'revenue': 'sum'}).reset_index()

romi = pd.merge(n_buyers, costs_2, on = ['source_id'])

romi = romi.merge(revenue, on = ['source_id'])

romi['cac'] = romi['costs'] / romi['n_buyers']
romi = romi.groupby(['source_id']).agg({'costs': 'sum', 'revenue': 'sum', 'n_buyers': 'sum'}).reset_index()
romi['gp'] = romi['revenue'] * margin_rate
romi['ltv'] = romi['gp'] / romi['n_buyers']
romi['cac'] = romi['costs'] / romi['n_buyers']
romi['romi'] = romi['ltv'] / romi['cac']
romi
```
